# Import libraries

```python
from reportlab.pdfgen import canvas
from pypdf import PdfReader
from langchain_text_splitters import RecursiveCharacterTextSplitter
import nltk
import pandas as pd

nltk.download('punkt')
nltk.download('punkt_tab')
```

# Create Policy PDF

```python
pdf_name = "employee_policy.pdf"

c = canvas.Canvas(pdf_name)

text = """
Leave Policy

Employees receive 12 casual leaves annually.
Employees receive 15 sick leaves annually.
Unused casual leaves cannot be carried forward.

Work From Home Policy

Employees may work from home twice per week.
Manager approval is required for additional remote work.

Travel Policy

Travel expenses are reimbursed within 30 days.
Original receipts must be submitted for reimbursement.

Medical Insurance Policy

All employees are covered under company medical insurance.
Dependent coverage is available for spouse and children.
"""

y = 800

for line in text.split("\n"):
    c.drawString(50, y, line)
    y -= 20

c.save()

print("PDF Created Successfully")
```

# Load PDF and Extract Text

```python
reader = PdfReader(pdf_name)

document_text = ""

for page in reader.pages:
    document_text += page.extract_text()

print("Original Document:\n")
print(document_text)
```

# Display Total Characters and Words

```python
total_characters = len(document_text)
total_words = len(document_text.split())

print("Total Characters:", total_characters)
print("Total Words:", total_words)
```

# **Task 2 : Fixed Size Chunking**

# Function for Fixed Chunking

```python
def fixed_chunking(text, chunk_size):

    chunks = []

    for i in range(0, len(text), chunk_size):
        chunks.append(text[i:i+chunk_size])

    return chunks
```

# Chunk Size = 100

```python
chunks_100 = fixed_chunking(document_text, 100)

for i, chunk in enumerate(chunks_100, 1):
    print(f"\nChunk {i}")
    print(chunk)
    print("Length:", len(chunk))
```

# Chunk Size = 200

```python
chunks_200 = fixed_chunking(document_text, 200)

for i, chunk in enumerate(chunks_200, 1):
    print(f"\nChunk {i}")
    print(chunk)
    print("Length:", len(chunk))
```

# Chunk Size = 300

```python
chunks_300 = fixed_chunking(document_text, 300)

for i, chunk in enumerate(chunks_300, 1):
    print(f"\nChunk {i}")
    print(chunk)
    print("Length:", len(chunk))
```


FIXED SIZE CHUNKING ANALYSIS

1. Are sentences being split?
   Yes, some sentences may be cut in the middle.

2. Is context preserved?
   Partially. Context can be lost when a sentence spans multiple chunks.

3. Limitations
   - Breaks sentence boundaries.
   - Can separate related information.
   - Retrieval quality may decrease.


# **Task 3 : Recursive Chunking**

# Recursive Chunking (200,20)

```python
splitter_1 = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20
)

chunks = splitter_1.split_text(document_text)

for i, chunk in enumerate(chunks, 1):
    print(f"\nChunk {i}")
    print(chunk)
```

# Recursive Chunking (200,50)

```python
splitter_2 = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=50
)

chunks = splitter_2.split_text(document_text)

for i, chunk in enumerate(chunks, 1):
    print(f"\nChunk {i}")
    print(chunk)
```

# Recursive Chunking (300,50)

```python
splitter_3 = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

chunks = splitter_3.split_text(document_text)

for i, chunk in enumerate(chunks, 1):
    print(f"\nChunk {i}")
    print(chunk)
```

RECURSIVE CHUNKING ANALYSIS

Overlap improves context retention because:

- Important information appears in adjacent chunks.
- Reduces loss of meaning near chunk boundaries.
- Improves retrieval accuracy.
- Better suited for RAG applications than fixed-size chunking.

# **Task 4 : Sentence Chunking**

# Sentence Chunking

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize(document_text)

for i, sentence in enumerate(sentences, 1):
    print(f"\nSentence {i}")
    print(sentence)
```


SENTENCE CHUNKING ANALYSIS

Yes, sentence chunking preserves meaning better than fixed-size chunking.

Reasons:
- Complete sentences remain intact.
- Better semantic understanding.
- More natural retrieval units.
- Reduced context loss.

# **Task 7 : Retrieval Simulation**

# Queries

```python
queries = [
    "How many casual leaves are provided?",
    "Can employees work from home?",
    "What is the travel reimbursement process?",
    "Who is covered under medical insurance?"
]
```

# Simple Retrieval Function

```python
def retrieve(query, chunks):

    query_words = set(query.lower().split())

    best_chunk = ""
    best_score = 0

    for chunk in chunks:

        score = sum(
            word in chunk.lower()
            for word in query_words
        )

        if score > best_score:
            best_score = score
            best_chunk = chunk

    return best_chunk
```

# Retrieval Using Fixed Size

```python
for query in queries:

    result = retrieve(query, chunks_200)

    print("\nQuery:", query)
    print("Chunking Method: Fixed Size")
    print("Retrieved Chunk:\n", result)
```

# Retrieval Using Recursive

```python
recursive_chunks = splitter_2.split_text(document_text)

for query in queries:

    result = retrieve(query, recursive_chunks)

    print("\nQuery:", query)
    print("Chunking Method: Recursive")
    print("Retrieved Chunk:\n", result)
```

# Retrieval Using Sentence Chunking

```python
for query in queries:

    result = retrieve(query, sentences)

    print("\nQuery:", query)
    print("Chunking Method: Sentence")
    print("Retrieved Chunk:\n", result)
```

# Retrieval Using Semantic Chunking

```python
semantic_list = list(semantic_chunks.values())

for query in queries:

    result = retrieve(query, semantic_list)

    print("\nQuery:", query)
    print("Chunking Method: Semantic")
    print("Retrieved Chunk:\n", result)
```
